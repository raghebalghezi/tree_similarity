```python
from main import SentGraph, Tree2graph
import matplotlib.pyplot as plt
import networkx as nx
```

```python
s1 = "(S (NP (D a) (JJ bad) (N researcher)) (VP (V shred) (NP (D the) (N paper))))"
s2 = "(S (NP (D the) (JJ angry) (N dog)) (VP (V ate) (NP (D the) (N steak))))"

score = SentGraph(s1, s2)

ent1 = Tree2graph(s1)
ent2 = Tree2graph(s2)
g1 = ent1.graph
g2 = ent2.graph
```

```python
print("Jaccard Similairty between S1 and S2 is {}".format(score.jaccard()))
print("Cosine Similairty between S1 and S2 is {}".format(score.cos_sim()))
```
